# Lung Cancer Prediction Model Using Support Vector Classification

## Imports
Numpy,
Pandas,
Sklearn &
Matplotlib

```python
import numpy as np
import pandas as pd

from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

from matplotlib import pyplot as plt
```

## Narrowing Down To Fewer Columns

```python
data = pd.read_csv("cancer patient data sets.csv")
data.pop('Patient Id')
data.pop('index')
data.pop('chronic Lung Disease')
data.pop('Balanced Diet')
data.pop('Air Pollution')
data.pop('Swallowing Difficulty')
data.pop('Fatigue')
data.pop('Clubbing of Finger Nails')
data.pop('Snoring')
data.pop('Coughing of Blood')
data.pop('OccuPational Hazards')
data.pop('Genetic Risk')
data.pop('Passive Smoker')
data.pop('Obesity')
data.pop('Frequent Cold')
data.pop('Weight Loss')

data
```

## Plotting Tabluar Data on X and y axis

```python
X = np.array(data.drop('Level', axis=1))
y = np.array(data['Level'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

## Creating the Classification Model & Making Predictions

```python
model = SVC()
model.fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)

for x in range(len(predictions)):
    print(y_test[x],"\t-\t", predictions[x])
```

## Calcluating & Displaying Accuracy

```python
accuracy = accuracy_score(y_test, predictions)
print("Model Accuracy",accuracy*100,"%")
```
